```python
# Seperating our indepent variables

x = df.iloc[:,:-1]
x
```

```python
# defining our dependent variablee

y = df.iloc[:,6]
y
```

Feature Engineering

```python
# convert categoritical values into numerical 

male_or_female = pd.get_dummies(x['sex'], drop_first=False)
```

```python
male_or_female
```

```python
smoker_or_nosmoker =pd.get_dummies(x['smoker'],drop_first = False)
```

```python
smoker_or_nosmoker
```

```python
region_ = pd.get_dummies(x['region'], drop_first=False)
region_
```

```python
#Drop sex categorical column from dataset
x=x.drop('sex', axis=1)
x[0:3]
```

```python
# similarly drop smoker and region
x = x.drop('smoker', axis=1)
x[0:3]
```

```python
# drop region column
x = x.drop('region', axis = 1)
x[0:3]
```

```python
# Concat the numerical forms of the csategorical columns 
x=pd.concat([x,male_or_female],axis=1)
x[0:3]
```

```python
x=pd.concat([x,smoker_or_nosmoker],axis=1)
x[0:10]
```

```python
x=pd.concat([x,region_],axis=1)
```

```python
x[0:5]
```

```python
#independent features = [age','bmi','children','female','male','no','yes','northeast','northwest','southeast','southwest']
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
model = scaler.fit(x)
x=model.transform(x)
print(x)
```

Train and Test Split

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

Model creation and Training

```python
#now create the model and train it over training data
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#now predict the values for the test data
y_pred = model.predict(x_test)
y_pred
```

Model Score

```python
from sklearn.metrics import r2_score # r2_score is used to compare the values which are predicted by model and actual values
                                        # so that we can find out accuracy of our model
score=r2_score(y_test,y_pred)
score
```
